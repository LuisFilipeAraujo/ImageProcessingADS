## **RESIZE**

```python
largura = imgrgb.shape[1] 
altura = imgrgb.shape[0]
```

```python
proporcao = float(altura/largura) 
largura_nova = 250 #em pixels 
altura_nova = int(largura_nova*proporcao)
tamanho_novo = (largura_nova, altura_nova)
print(tamanho_novo)
```

```python
img_redimensionada = cv2.resize(imgrgb, tamanho_novo, interpolation = cv2.INTER_AREA)
```

```python
plt.imshow(img_redimensionada)
plt.title("Imagem redimensionada utilizando a função resize")
```

## **Rotação**

```python
Altura = (img_redimensionada2.shape[0])
Largura = (img_redimensionada2.shape[1])
Canais = (img_redimensionada2.shape[2])
```

```python
print("Altura (height): %d pixels" %Altura)  
print("Largura (width): %d pixels" %Largura) 
print("Canais (channels): %d" %Canais)
```

```python
centro = (Largura // 2, Altura // 2)
centro
```

```python
M = cv2.getRotationMatrix2D(centro, 90, 1.0) #90 graus
```

```python
img_rotacionada = cv2.warpAffine(img_redimensionada2, M, (Largura, Altura))
```

```python
plt.imshow(img_rotacionada)
plt.title("Imagem Rotacionada")
plt.show()
```
